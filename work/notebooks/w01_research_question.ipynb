## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*
Looking at the starter dataset, we have an inventory of 30000 pages. Out of those, 16262 are currently marked with a downward trend. More importantly, 8031 of those declining pages have significant search volume (over 1,000 impressions in the last 90 days). This proves there is a substantial, filterable pool of high-value targets that warrants a structured ranking system over the next 7 weeks.

```python
import pandas as pd

#Using the Raw GitHub URL to bypass Colab's local file storage (fixed it after an error was raised)
csv_url = 'https://raw.githubusercontent.com/somyamaheshwari2612/flyyyyy/main/data/raw/content_refresh_anonymized.csv'

#Loading the small anonymized starter dataset directly from the web
df = pd.read_csv(csv_url)

#Calculating a few quick baseline metrics to prove there is signal to evaluate
total_pages = len(df)
declining_pages = len(df[df['trend_direction'] == 'down'])
high_visibility_declining = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] > 1000)])

print(f"Total pages in the starter dataset: {total_pages}")
print(f"Pages showing a downward trend: {declining_pages}")
print(f"Declining pages with significant volume (>1000 impressions in 90 days): {high_visibility_declining}")
```
