## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# Your discovery here
#Discovery A
df_active = df[df["impressions_90d"]>0] #Filtering df
corr = df_active["search_volume"].corr(df_active["impressions_90d"]) #Calculating Correlation
print(f"Correlation between search_volume and impressions_90d (Active Pages): {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a ranking page actually gets.")
```

```python
#Discovery B
#Grouping by position_tier and content_type
ctr_by_pos_and_type = visible.groupby(["position_tier", "content_type"])["ctr"].mean().reset_index()
sorted_ctr = ctr_by_pos_and_type.sort_values(by=["position_tier", "ctr"], ascending=[True, True]) #Sorting to make sure lowest CTRs are at top
worst_performers = sorted_ctr.drop_duplicates(subset=["position_tier"], keep="first") #Extracting row with lowest CTR for each position_tier
print("Content Types with the Worst CTR per Position Tier:")
print(worst_performers.to_string(index=False))
pivot_df = visible.pivot_table(index="position_tier", columns="content_type", values="ctr", aggfunc="mean") #Creating a pivot table for easy plotting
pivot_df.plot(kind="bar", figsize=(10, 6), title="Mean CTR by Position Tier and Content Type", ylabel="CTR") #Plotting a grouped bar chart
```

```python
#Additional Discovery
#Does content_age_days relate to trend_direction
#The Hypothesis: Do older articles tend to decay (trend downward), while newer articles grow (trend upward)?
#Assuming trend_direction is a categorical column (like "Up", "Down", "Flat") and content_age_days is numeric, the best way to test this is by comparing the median age across the different trend buckets, and visualizing it with a boxplot.
import seaborn as sns
import matplotlib.pyplot as plt

#Getting the median age for each trend direction to see the baseline
age_by_trend = df.groupby("trend_direction")["content_age_days"].median()
print("Median Content Age by Trend Direction:")
print(age_by_trend)
print("-" * 30)

#Visualizing the spread to see if there are clear patterns
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="trend_direction", y="content_age_days")
plt.title("Does Content Age Impact Traffic Trends?")
plt.ylabel("Content Age (Days)")
plt.xlabel("Trend Direction")
plt.show()
#My hypothesis was likely that older content is rotting and trending downward. The data shows the exact opposite.
#Trending Down / Flat: ~7 to 7.5 months old (216–231 days)
#Trending Up / Stable: ~9.5 to 10 months old (291–300 days)
```

```python
#Does avg_position relate to ctr?
#The Hypothesis: Better rankings (lower position numbers) yield higher click-through rates.
#Ensuring we are only looking at pages that actually get impressions
visible = df[df["impressions_90d"] > 0]

#Calculating Spearman correlation (measures non-linear relationships)
#Expecting a strong NEGATIVE correlation (as position goes UP from 1 to 10, CTR goes DOWN)
pos_ctr_corr = visible["avg_position"].corr(visible["ctr"], method="spearman")
print(f"Spearman Correlation between Average Position and CTR: {pos_ctr_corr:.3f}")
print("Closer to -1.0 means a perfectly predictable CTR drop-off as ranking worsens.")

#Visualizing the classic CTR curve
visible.plot.scatter(
    x="avg_position",
    y="ctr",
    alpha=0.4, #Making dots transparent to see clusters
    figsize=(10, 6),
    title="The CTR Decay Curve: Avg Position vs. CTR",
    xlabel="Average Position (Lower is Better)",
    ylabel="Click-Through Rate (CTR)"
)

#Limiting the X-axis to the top 20 positions where the real action happens
plt.xlim(0, 20)
plt.show()
#A score of -0.144 means the traditional "click curve" is fundamentally broken on the site.
```
